```python
import sympy
```

```python
x = sympy.Symbol("x")
```

```python
formula = sympy.sin(x)**2 - sympy.cos(x)**2
formula
```

```python
#1. Определить корни (найти такие х при которых f(x) = 0)
formula_solve = sympy.solve(formula)
formula_solve
```

```python
#2. Найти интервалы, на которых функция возрастает (f'(x) > 0)
exp = sympy.diff(formula)
exp
```

```python
up_exp = sympy.solve(exp > 0)
up_exp
```

```python
#3. Найти интервалы, на которых функция убывает (f'(x) < 0)
down_exp = sympy.solve(exp < 0)
down_exp
```
